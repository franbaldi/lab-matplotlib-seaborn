#### From the Fitbit data, we want to visually understand:

```python
# MEDIAN STEPS BY MONTH_ENCODED
fitbit[fitbit['Months_encoded']==1].mean()["Steps"]
```

```python
fig, [f1, f2] = plt.subplots(1,2, figsize=(15,5), sharey=True, sharex=False)
bins_f1 = ['Jan', 'Feb', 'Mar', 'Apr', 'May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
f1.bar(bins_f1, fitbit.groupby(by = 'Months_encoded').median()["Steps"], label="Montly avg Steps")
f1.set_title('Montly avg Steps', fontsize=12)   
f1.set_xlabel('months', fontsize=12)     
f1.set_ylabel('avg steps', fontsize=12)
f1.legend(fontsize=14)

# MEDIAN STEPS BY WORK_OR_WEEKEND
bins_f2 = ['Weekend','Work Days']
f2.bar(bins_f2, fitbit.groupby(by = 'Work_or_Weekend').median()["Steps"], label="Montly avg Steps")
f2.set_title('Work_Vs_Weekend', fontsize=12)   
f2.set_xlabel('Work_or_Weekend', fontsize=12)     
f2.set_ylabel('avg steps', fontsize=12)
f2.legend(fontsize=14)
```

```python
plt.subplots_adjust(wspace=1)
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

variables = ["Minutes Lightly Active", "Minutes Very Active", "Minutes Sedentary"]
for i in variables:
    plt.scatter(fitbit["Steps"], fitbit[f"{i}"], label=f"{i} vs. Steps per day")

    plt.xlabel("Steps")
    plt.ylabel(f"{i}")
    plt.title(f"{i} vs. Steps per day")
    plt.legend()
    plt.show()
```

#### Plot the correlation matrix using seaborn.

```python
# your code here
corr = titanic.corr()

# Generate a custom diverging colormap
cmap = sns.diverging_palette(300, 200, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, cmap=cmap, vmax=1, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

#### What are the most correlated features?
